## Outlier Detection Using Boxplots

Outlier analysis helps identify unusually high or low values in the dataset.  
In this task, boxplots are used to visualize outliers in sales and profit columns.

Outliers may represent:
- Bulk customer orders
- High-value transactions
- Extremely low or negative profits

These insights are important for future business analysis and dashboarding.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['sales'])

plt.title("Sales Outlier Analysis")
plt.xlabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['profit'])

plt.title("Profit Outlier Analysis")
plt.xlabel("Profit")

plt.show()
```

### Observation

The boxplots show the presence of outliers in both sales and profit columns.  
Some transactions have exceptionally high sales and profits, while a few records show significant losses.

These outliers may represent:
- Large corporate purchases
- Heavy discount transactions
- Loss-making sales

The outliers were observed and retained because they may contain valuable business information.
